```python
def rating(value):
    value=str(value)
    
    if '/' in value:
        value=str(value).split("/")
        value=value[0]
        return float(value)

```

```python
df['rate']=df['rate'].apply(rating)
df['rate'].head()
```

```python
def price(value):
    value=str(value)
    value=value.replace(",","")
    return float(value)
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].apply(price)
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)']//2
```

```python
def delivery(value):
    value=str(value)
    if value=="Yes":
        return 1
    else :
        return 0
```

```python
area_distribution = df['location'].value_counts()
cheapest_per_cuisine = df.groupby('cuisines')['cost of one person'].min()
expensive_per_cuisine = df.groupby('cuisines')['cost of one person'].max()
```

```python
max_delivery_reviews_location = df[df['votes'] > 1000]['location'].value_counts().idxmax()
interesting_insight = df[['rate', 'votes']].corr().loc['rate', 'votes']
max_less_rated_location = df[df['rate'] < 3.0]['location'].value_counts().idxmax()
```

```python
cheap_expensive_avg_price = df.groupby('location')['cost of one person'].agg(['min', 'max', 'mean'])
num_restaurants_per_cuisine = df['cuisines'].value_counts()
```

```python
app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='Restaurant Insights Dashboard'),
    
    html.Div(children='''
        Area-wise Distribution of Restaurants
    '''),
    dcc.Graph(
        id='area-distribution',
        figure={
            'data': [
                {'x': area_distribution.index, 'y': area_distribution.values, 'type': 'bar'}
            ],
            'layout': {
                'title': 'Area-wise Distribution of Restaurants'
            }
        }
    ),
    
    html.Div(children='''
        Cheapest and Most Expensive Restaurant for Each Cuisine
    '''),
    html.Table([
        html.Thead([
            html.Tr([html.Th('Cuisine'), html.Th('Cheapest Restaurant'), html.Th('Most Expensive Restaurant')])
        ]),
        html.Tbody([
            html.Tr([html.Td(cuisine), html.Td(cheapest_per_cuisine[cuisine]), html.Td(expensive_per_cuisine[cuisine])])
            for cuisine in cheapest_per_cuisine.index
        ])
    ]),
    
    html.Div(children='''
        Location with Maximum Number of Restaurants with Review Votes > 1000: {}
    '''.format(max_delivery_reviews_location)),
    
    html.Div(children='''
        Interesting Insight: Correlation between Rate and Votes: {:.2f}
    '''.format(interesting_insight)),
    
    html.Div(children='''
        Location with Maximum Number of Less-Rated Restaurants: {}
    '''.format(max_less_rated_location)),
    
    html.Div(children='''
        Area-wise Cheap, Expensive, and Average Prices:
    '''),
    html.Table([
        html.Thead([
            html.Tr([html.Th('Location'), html.Th('Min Cost'), html.Th('Max Cost'), html.Th('Avg Cost')])
        ]),
        html.Tbody([
            html.Tr([html.Td(location), html.Td(cheap_expensive_avg_price.loc[location, 'min']),
                     html.Td(cheap_expensive_avg_price.loc[location, 'max']), html.Td(cheap_expensive_avg_price.loc[location, 'mean'])])
            for location in cheap_expensive_avg_price.index
        ])
    ]),
    
    html.Div(children='''
        Number of Restaurants for Each Type of Cuisine
    '''),
    dcc.Graph(
        id='cuisine-count',
        figure={
            'data': [
                {'x': num_restaurants_per_cuisine.index, 'y': num_restaurants_per_cuisine.values, 'type': 'bar'}
            ],
            'layout': {
                'title': 'Number of Restaurants for Each Type of Cuisine'
            }
        }
    ),
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
